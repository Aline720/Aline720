```python
# Calculadora de IMC - Versão 2026

def calcular_imc():
    print("--- Calculadora de Índice de Massa Corporal ---")

    try:
        # Entrada de dados convertida para decimal (float)
        peso = float(input("Digite seu peso em kg (ex: 70.5): "))
        altura = float(input("Digite sua altura em metros (ex: 1.75): "))

        # Fórmula: IMC = peso / altura²
        imc = peso / (altura ** 2)

        print(f"\nSeu IMC é: {imc:.2f}")

        # Classificação baseada na tabela da OMS (Jan/2026)
        if imc < 18.5:
            resultado = "Abaixo do peso"
        elif 18.5 <= imc < 25:
            resultado = "Peso normal"
        elif 25 <= imc < 30:
            resultado = "Sobrepeso (Pré-obesidade)"
        elif 30 <= imc < 35:
            resultado = "Obesidade Grau I"
        elif 35 <= imc < 40:
            resultado = "Obesidade Grau II"
        else:
            resultado = "Obesidade Grau III (Mórbida)"

        print(f"Classificação: {resultado}")

    except ValueError:
        print("Erro: Use apenas números e utilize o ponto (.) como separador decimal.")
    except ZeroDivisionError:
        print("Erro: A altura não pode ser zero.")

if __name__ == "__main__":
    calcular_imc()
```
